```python
with open("../data/aa_freq_byPDB.tsv", 'r') as fh:
    fh.readline()
    
    count = {}
    count['chain'] = {}
    count['chain']['residue'] = {}
    
    for(line) in fh:
        fields = line.split()
        pdb = fields[0]
        
        if(pdb != "6fml" and pdb != "6etx"):
            chain = fields[1]
            res = fields[2]
            i = int(fields[3])
            
            if(chain in count.keys()):
                if(res in count[chain].keys()):
                    count[chain][res] += i
                else:
                    count[chain][res] = i
            else:
                count[chain] = {res : i}
                
    for(chain) in count.keys():
                
        for(res) in count[chain].keys():
            print(chain+'\t'+res+'\t', count[chain][res])
```
