```python
def map_chains(labeledChainsFile):
    with open("../data/labeled_chains2.tsv", 'r') as lch:
        chainDict = {}
        chainDict['pdbID'] = {} 
        chainDict['pdbID']['chain'] = {}
        #chainDict['pdbID']['chain']['uniprot'] = {}
        
        lch.readline()
        for(line) in lch:
            fields = line.split('\t')
            pdbID = fields[0].lower()
            chain = fields[1]
            function = fields[5]
            uniprot = fields[6]
            name = fields[7]
            
            if(pdbID in chainDict):
                if(chain not in chainDict[pdbID]):
                    chainDict[pdbID][chain] = uniprot + ',' + name + ',' + function
            else:
                chainDict[pdbID] = {chain : uniprot + ',' + name + ',' + function}
                
        return chainDict
```

```python
with open("../data/aa_freqByPDB&BP.tsv", 'r') as afh:
    count = {}
    count['source'] = {}
    count['source']['target'] = {}
    count['source']['target']['residue'] = {}
    
    norm = {}
    norm['source'] = {}
    norm['source']['target'] = {}
    norm['source']['target']['residue'] = {}

    chainDict = map_chains("../data/labeled_chains2.tsv")

    afh.readline()
    for line in afh:
        fields = line.split('\t')
        pdbID = fields[0]
        sourceChain = fields[1][0]
        targetChain = fields[1][1]
        res = fields[2]
        atomContacts = int(fields[3])

        sourceUNIPROT = chainDict[pdbID][sourceChain].split(',')[0]
        targetUNIPROT = chainDict[pdbID][targetChain].split(',')[0]

        if(sourceUNIPROT in count):
            print(sourceUNIPROT + ' is in count')
            if(targetUNIPROT in count[sourceUNIPROT]):
                print(targetUNIPROT + ' is in count ' + sourceUNIPROT)
                if(res in count[sourceUNIPROT][targetUNIPROT]):
                    print(res + ' is in count ' + sourceUNIPROT + ' ' + targetUNIPROT)
                    count[sourceUNIPROT][targetUNIPROT][res] += atomContacts
                    norm[sourceUNIPROT][targetUNIPROT][res] += 1
                    print(count)
                else:
                    print('adding res')
                    count[sourceUNIPROT][targetUNIPROT] = {res : atomContacts}
                    norm[sourceUNIPROT][targetUNIPROT] = {res : 1}
                    print(count)
            else:
                print('adding targetUNIPROT')
                count[sourceUNIPROT] = {targetUNIPROT : {res : atomContacts}}
                norm[sourceUNIPROT] = {targetUNIPROT : {res : 1}}
                print(count)
        else:
            print('adding sourceUNIPROT')
            count = {sourceUNIPROT : {targetUNIPROT : {res : atomContacts}}}
            norm = {sourceUNIPROT : {targetUNIPROT : {res : 1}}}
            print(count)

    for source in count:
        for target in count[source]:
            for res in count[source][target]:
                print(source + '\t' + target + '\t' + res + '\t' + str(count[source][target][res]) + 't' + str(norm[source][target][res]))
```
